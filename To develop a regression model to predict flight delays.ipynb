# Model Training:

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score
```

```python
x=df.drop(['DEP_DELAY'],axis=1)
x
```

```python
y=df["DEP_DELAY"]
y
```

```python
x_train,x_test,y_train, y_test = train_test_split(x, y, test_size=0.20, random_state=2)
```

```python
df.shape
```

```python
x_train.shape ,x_test.shape
```

```python
y_train.shape ,y_test.shape
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler=StandardScaler()
```

```python
x_train= scaler.fit_transform(x_train)
```

```python
x_test= scaler.transform(x_test)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
regression=LinearRegression()
```

```python
regression.fit(x_train,y_train)
```

```python
y_hat= regression.predict(x_test)
```

```python
y_hat
```

```python
# Evaluation of the model:

mse = mean_squared_error(y_test, y_hat)
print(mse)
```

```python
rmse = np.sqrt(mse)
print(rmse)
```

```python
Accuracy= r2_score(y_test, y_hat)
print(Accuracy)
```

```python
# Visualize the results

plt.figure(figsize=(10, 6))
plt.scatter(y_test, y_hat)
plt.xlabel('Actual Delay')
plt.ylabel('Predicted Delay')
plt.title('Actual vs Predicted Delay')
plt.show()
```

```python
# Trying with another algorithm:

from sklearn.svm import SVR
svr = SVR(kernel ='rbf')

svr.fit(x_train, y_train)

# prediction
y_hat = svr.predict(x_test)

# Evaluate the performance
from sklearn.metrics import r2_score
Accuracyy = r2_score(y_test, y_hat)
print(Accuracyy)
```

```python
from sklearn.ensemble import RandomForestRegressor
rfr = RandomForestRegressor(n_estimators=550,random_state=22,min_samples_leaf=.0001)
rfr.fit(x_train,y_train)

# Prediction
y_hat =rfr.predict(x_test)

# Evaluate the performance
from sklearn.metrics import r2_score

print(r2_score(y_test,y_hat))
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(y_test, y_hat)
plt.xlabel('Actual Delay')
plt.ylabel('Predicted Delay')
plt.title('Actual vs Predicted Delay')
plt.show()
```
